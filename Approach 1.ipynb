```python
# Operation intervals of vehicles
df = pd.read_csv('4_einsaetze_land.csv', sep=',')
df
```

```python
columns = ['land','ort','freies_merkmal','fuehrerscheinklasse','einsatzort', 'kostenstelle']
df.drop(columns, inplace=True, axis=1)
```

```python
df = df.sort_values('equipment_nummer')
df['equipment_nummer'].nunique()
```

```python
df
```

```python
df['einsatzbeginn'] = pd.to_datetime(df['einsatzbeginn'], dayfirst=True)
df['einsatzbeginn'] = pd.to_datetime(df['einsatzbeginn'], format='%d/%m/%y %H:%M:%S')
df['einsatzende'] = pd.to_datetime(df['einsatzende'], dayfirst=True)
df['einsatzende'] = pd.to_datetime(df['einsatzende'], format='%d/%m/%y %H:%M:%S')
```

```python
df['operated_hours'] = (df.einsatzende-df.einsatzbeginn) / pd.Timedelta(hours=1)
```

```python
df
```

```python
df['Total_operated_hours_per_vehicle'] = df.groupby('equipment_nummer')['operated_hours'].cumsum()
```

```python
df = df.drop_duplicates(subset='equipment_nummer', keep="last")
```

```python
df["segment"] = df["segment"].astype('category')
df["transpondertyp"] = df["transpondertyp"].astype('category')
df["schichttyp"] = df["schichttyp"].astype('category')
df["logout"] = df["logout"].astype('category')
```

```python
columns = ['einsatzbeginn','einsatzende','warenempfaenger','interne_nummer', 'warenempfaenger_nummer']
df.drop(columns, inplace=True, axis=1)
```

```python
df = df.dropna()
```

```python
df
```

```python
# Encoding non float or integer variables
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OrdinalEncoder

df_encoded = df.copy()

ord_enc = OrdinalEncoder()
df_encoded["segment"] = ord_enc.fit_transform(df_encoded[["segment"]])
df_encoded["transpondertyp"] = ord_enc.fit_transform(df_encoded[["transpondertyp"]])
df_encoded["schichttyp"] = ord_enc.fit_transform(df_encoded[["schichttyp"]])
df_encoded["logout"] = ord_enc.fit_transform(df_encoded[["logout"]])
df_encoded.head()
```

```python
df_encoded.info()
```

```python
df_encoded.corr()
```

```python
training_df = df_encoded.sample(frac=0.7,random_state=200) 
test_df = df_encoded.drop(training_df.index)
```

```python
Real_operated_hours = test_df["Total_operated_hours_per_vehicle"]
Total_operated_hours_per_vehicle = training_df["Total_operated_hours_per_vehicle"]
training_df = training_df.drop('Total_operated_hours_per_vehicle', axis=1)
test_df = test_df.drop('Total_operated_hours_per_vehicle', axis=1)
```

```python
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(max_depth=8)
modelRFR = regressor.fit(training_df, Total_operated_hours_per_vehicle)

res = modelRFR.predict(test_df)
predicted_df = pd.Series(res)

from sklearn.metrics import r2_score
r_square = modelRFR.score(training_df, Total_operated_hours_per_vehicle)
print("Random Forest Regressor R-squared on training data: " + str(r_square))
```

```python
from sklearn.metrics import r2_score
r_square = modelRFR.score(test_df, Real_operated_hours)
print("Random Forest Regressor R-squared on test data: " + str(r_square))
```

```python
from sklearn.metrics import mean_squared_error

rms = np.sqrt(mean_squared_error(res, Real_operated_hours))
print(rms)
```
