```python
# Operation intervals of vehicles
df = pd.read_csv('4_einsaetze_land.csv', sep=';')
df
```

```python
columns = ['land','ort','freies_merkmal','fuehrerscheinklasse','einsatzort', 'kostenstelle']
df.drop(columns, inplace=True, axis=1)
```

```python
df = df.sort_values('equipment_nummer')
df['equipment_nummer'].nunique()
```

```python
df = df.drop_duplicates()
```

```python
df = df.dropna()
```

```python
df['einsatzbeginn'] = pd.to_datetime(df['einsatzbeginn'], dayfirst=True)
df['einsatzbeginn'] = pd.to_datetime(df['einsatzbeginn'], format='%d/%m/%y %H:%M:%S')
df['einsatzende'] = pd.to_datetime(df['einsatzende'], dayfirst=True)
df['einsatzende'] = pd.to_datetime(df['einsatzende'], format='%d/%m/%y %H:%M:%S')
```

```python
df['operated_minutes'] = (df.einsatzende-df.einsatzbeginn) / pd.Timedelta(minutes=1)
```

```python
df['operated_minutes'] = df['operated_minutes'].round(decimals=2)
```

```python
df
```

```python
df.info()
```

```python
df["segment"] = df["segment"].astype('category')
df["transpondertyp"] = df["transpondertyp"].astype('category')
df["schichttyp"] = df["schichttyp"].astype('category')
df["logout"] = df["logout"].astype('category')
```

```python
df.info()
```

```python
columns = ['einsatzbeginn','einsatzende','warenempfaenger','interne_nummer', 'warenempfaenger_nummer']
df.drop(columns, inplace=True, axis=1)
```

```python
df.info()
```

```python
# Encoding non float or integer variables
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OrdinalEncoder

df_encoded = df.copy()

ord_enc = OrdinalEncoder()
df_encoded["segment"] = ord_enc.fit_transform(df_encoded[["segment"]])
df_encoded["transpondertyp"] = ord_enc.fit_transform(df_encoded[["transpondertyp"]])
df_encoded["schichttyp"] = ord_enc.fit_transform(df_encoded[["schichttyp"]])
df_encoded["logout"] = ord_enc.fit_transform(df_encoded[["logout"]])
df_encoded.head()
```

```python
import math
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import seed
seed(1)
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf

import tensorflow
tensorflow.random.set_seed(1)
from tensorflow.keras import layers
from tensorflow.keras import activations
from tensorflow.keras.layers import Dropout
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.wrappers.scikit_learn import KerasRegressor

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
training_df = df_encoded.sample(frac=0.7,random_state=200) 
test_df = df_encoded.drop(training_df.index)
```

```python
operated_minutes = training_df["operated_minutes"]
training_df = training_df.drop('operated_minutes', axis=1)
test_df = test_df.drop('operated_minutes', axis=1)
```

```python
from sklearn.neural_network import MLPRegressor

modelMLP = MLPRegressor(hidden_layer_sizes=(21), max_iter=1000)
modelMLP.fit(training_df, operated_minutes)

res = modelMLP.predict(test_df)
predicted_df = pd.Series(res)

from sklearn.metrics import r2_score
r_square = modelMLP.score(training_df, operated_minutes)
print("Score Multi-layer Perceptron Regressor: " + str(r_square))
```

```python
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(max_depth=16)
modelRFR = regressor.fit(training_df, operated_minutes)

res = modelRFR.predict(test_df)
predicted_df = pd.Series(res)

from sklearn.metrics import r2_score
r_square = modelRFR.score(training_df, operated_minutes)
print("Score Random Forest Regressor: " + str(r_square))
```

```python
from sklearn.tree import DecisionTreeRegressor

regressor = DecisionTreeRegressor(max_depth=16)
modelDTR = regressor.fit(training_df, operated_minutes)

res = modelDTR.predict(test_df)
predicted_df = pd.Series(res)

from sklearn.metrics import r2_score
r_square = modelDTR.score(training_df, operated_minutes)
print("Score Decision Tree Regressor: " + str(r_square))
```
